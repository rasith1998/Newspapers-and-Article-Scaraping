### Working with Newspapers

```python
url = input("Enter newspaper URL: ")
```

```python
import newspaper
newspaper = newspaper.build(url, memoize_articles=False) ##, language='zh') #memoize_articles=False is for not eleminate
```

```python
newspaper.size()# No of news articles
```

```python
newspaper.brand # News owner
```

```python
newspaper.description
```

```python
articles = newspaper.articles # articles present in the newspaper
for article in articles:
    print(article.url)
```

```python
categories = newspaper.category_urls() # URLs for categories which belongs to the newspaper
```

```python
for category in categories:
    print(category)
```

```python
feeds = newspaper.feed_urls() # URLs for feeds which available in the newspaper
```

```python
for feed in feeds:
    print(feed)
```

```python
article = articles[10]

article.download()
article.parse()
article.html
```

## HTML to Plain Text

```python
url = input("Enter any URL: ")
response = requests.get(url) 
html = response.text 
plainText = fulltext(html)
plainText
```
